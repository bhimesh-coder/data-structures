## Data Structures
1. What are data structures, and why are they important?

Data structures are ways of organizing and storing data so that it can be accessed and modified efficiently. They are important because they help in optimizing performance and managing data logically and efficiently.

2. Difference between mutable and immutable data types with examples

Mutable: Can be changed after creation.
Example: list, dict, set

Immutable: Cannot be changed after creation.
Example: int, str, tuple


3. Differences between lists and tuples in Python:

List	Tuple
Mutable	Immutable
Slower	Faster
Uses []	Uses ()
Can be modified	Cannot be modified

4. How dictionaries store data:

Dictionaries store data as key-value pairs using a hash table. Keys are hashed to unique integers that determine where the value is stored in memory.

5. Why use a set instead of a list in Python?

To store unique elements (no duplicates)

Faster membership testing (in)

Ideal for set operations like union, intersection

6. What is a string in Python, and how is it different from a list?

A string is an immutable sequence of characters.
Difference: Strings are immutable and only contain characters, while lists are mutable and can contain any data type.

7. How do tuples ensure data integrity in Python?

Since tuples are immutable, their content cannot be changed, which makes them ideal for storing constant data and ensuring it remains unchanged.

8. What is a hash table, and how does it relate to dictionaries in Python?

A hash table is a data structure that maps keys to values using a hash function. Python dictionaries are implemented using hash tables.

9. Can lists contain different data types in Python?

Yes, lists in Python can store multiple data types in a single list.

10. Why are strings immutable in Python?

Immutability makes strings hashable, efficient for memory usage, and safe for concurrent access.

11. Advantages dictionaries offer over lists:
Faster lookup (O(1) average case)

Named access using keys

Suitable for key-value mappings

12. Scenario where using a tuple is preferable over a list:
When storing fixed data that should not change, e.g., coordinates ((x, y)), or passing as a key in a dictionary.

13. How do sets handle duplicate values in Python?
Sets automatically remove duplicates. Only unique elements are stored.

14. How the “in” keyword works differently:
List: Checks for value presence.

Dictionary: Checks for key presence.

15. Can you modify elements of a tuple?
No, tuples are immutable. Once created, elements cannot be changed.

16. What is a nested dictionary and its use case?
A dictionary inside another dictionary.

17. Time complexity of accessing elements in a dictionary:
Average case: O(1)
Worst case: O(n) (due to hash collisions)

18. Situations where lists are preferred over dictionaries:
When order matters

When only values are needed

For indexed access

19. Why dictionaries are unordered and how it affects retrieval:
Dictionaries (before Python 3.7) do not maintain order, which means order-based operations are unreliable. From Python 3.7+, insertion order is preserved but still not intended for sorting.

20. Difference between a list and a dictionary in data retrieval:
List: Access by index (position-based)

Dictionary: Access by key (key-based, more descriptive and efficient)
